## Modifying data

```python
d2 = df.copy()
# d2["Total"] = d2.HP + d2.Attack + d2.Defense
d2["Total"] = d2.iloc[:, 4:10].sum(axis=1)
d2.head()
```

```python
d3 = d2.drop(columns=["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"])
d3.head()
```

```python
cols = list(d3.columns.values)
d4 = d3[[cols[1]] + [cols[-1]] + cols[2:6]]
d4.head()
```

```python
d4.columns.get_loc("Type 1")
```

## Saving data

```python
d4.to_csv("modified.csv", index=False, sep="\t")
```

```python
d4.to_excel("modified.xlsx", index=False)
```

## Filtering data

```python
d5 = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]
d5.tail()
```

```python
d5.reset_index().head()
```

```python
d5 = d5.reset_index(drop=True)
d5.head()
```

```python
d6 = d4.loc[d4["Name"].str.contains("Mega")]
d6.head()
```

```python
d6 = d4.loc[~d4["Name"].str.contains("Mega")]
d6.head()
```

```python
import re

d6 = d4.loc[d4["Type 1"].str.contains("fire|grass", flags=re.I, regex=True)]
d6.head()
```

```python
d6 = d4.loc[df.Name.str.contains("^pi[a-z]*", flags=re.I, regex=True)]
d6.head()
```

## Aggregate Statistics

```python
da = df.copy()
da.head()
```

```python
da.groupby(["Type 1"]).mean().head()
```

```python
da.groupby(["Type 1"]).sum().head()
```

```python
da["count"] = 1
da.groupby(["Type 1"]).count()["count"].head()
```

```python
da.groupby(["Type 1"]).mean().sort_values("HP", ascending=False).head()
```

```python
da.groupby(["Type 1", "Type 2"]).count()["count"].head(16)
```
