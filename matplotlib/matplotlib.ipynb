## User guide

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,3,2,4])
```

```python
plt.plot([1,2,3,4], [1,3,2,4])
```

```python
fig, axs = plt.subplots(1,2)
a, b = axs

a.set_title("I'm A axe")
a.set_xlabel("We are Xs")
a.set_ylabel("We are Ys")

b.set_title("And I'm B axe")
```

```python
x = np.linspace(0,2)

fig, ax = plt.subplots()
ax.plot(x,x, label="linear")
ax.plot(x,x**2, label="quadratic")
ax.plot(x, -x+3, label="special")
ax.set_title("Awesome OO-style")
ax.legend()
```

```python
x = np.linspace(0,2,5)

plt.plot(x, x, label="linear")
plt.plot(x, x**2, label="quadratic")
plt.plot(x, -x+3, label="special")
plt.title("Awesome pyplot-style")
plt.legend()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
        ...
    """
    return ax.plot(data1, data2, **param_dict)

data1, data2 = np.random.randn(2,100)
fig, ax = plt.subplots()
my_plotter(ax, data1, data2, {"marker":"o"})
```

```python
import matplotlib as mpl

data1, data2 = np.random.randn(2,100)
fig, ax = plt.subplots()
my_plotter(ax, data1, data2, {"marker":"o"})

mpl.use("svg")
plt.savefig("plot")
```

## pyplot tutorial

```python
plt.plot([1,10,-10,0], [1,10,1,2])
```

```python
plt.plot([1,10,1,2])
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'mo')
plt.axis([0, 6, 0, 20])
```

```python
t = np.arange(0., 10., 0.4)
plt.plot(t, t, "r--", t, t**2, "bs", t, np.sin(t)*10+20, 'g^')
plt.axis([0, 10, 0, 60])
```

```python
data = { "a":np.random.randn(1000), "b":np.random.randn(1000), "c":np.random.randn(1000), "d":np.random.randn(1000)*20}
plt.scatter("a", "b", c="c", s="d", data=data)
```

```python
names = ["A", "B", "C"]
values = [1, 10, 100]

plt.figure(figsize=(14,4))

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values, c="r")

plt.subplot(133)
plt.plot(names, values, "y")
```

```python
plt.plot(np.random.rand(20), linewidth=1., c="k", linestyle="dotted", markersize=3, marker="o")
```

```python
plt.plot("Weight", "Height", data={"Weight":[1,3,5,9], "Height":[2,5,9,9.5]})
plt.xlabel("Weight")
plt.ylabel("Height")
```

```python
plt.plot([1,3,6,8,12,15,17,18,20])
plt.plot([4,5,9,11,14,17,17,18,19])
```

```python
t = np.arange(-10,10,0.1)

plt.subplot(211)
plt.plot(t, 1/(1 + np.exp(-t)), 'r--')
plt.axis([-10,10,-0.1,1.1])

plt.subplot(2,1,2)
plt.plot(t, t*0.05+0.5, 'b--')
plt.axis([-10,10,-0.1,1.1])
```

```python
x = np.random.randn(1000) * 100
```

```python
n, bins, patches = plt.hist(x, 20, density=True, facecolor='b', alpha=0.5)

plt.title("Histogram")
plt.grid(True)
plt.text(-250, 0.0036, "What's up?")
```

```python
n, bins, patches = plt.hist(x, 20, density=True, facecolor='g', alpha=0.5)
plt.grid(True)
plt.annotate("This is an annotation!", xy=(300, 0.0001), xytext=(-100,0.0015), arrowprops=dict(facecolor="k", shrink=1))
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.scatter(np.random.randn(100), np.random.randn(100), np.random.randn(100), "ro")
```
